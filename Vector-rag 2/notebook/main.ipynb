```python
import pytesseract
import cv2
from PIL import Image

def test_ocr_on_image(image_path):
    print(f"--- Testing OCR on: {image_path} ---")
    
    # 1. تحميل الصورة باستخدام OpenCV لعمل معالجة بسيطة
    image = cv2.imread(image_path)
    
    # تحويل الصورة لرمادي (بيحسن القراءة جداً)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # 2. تشغيل Tesseract على الصورة
    # config ده بيخلي الـ OCR يركز إنه يقرأ "كتلة نصية" واحدة (مثالي للروايات)
    custom_config = r'--oem 3 --psm 3' 
    text = pytesseract.image_to_string(gray, lang='eng', config=custom_config)
    
    # 3. طباعة النتيجة
    print("\n[EXTRACTED TEXT]:")
    print("-" * 30)
    if text.strip():
        print(text)
    else:
        print("للأسف مطلعش نص! تأكد من جودة الصورة أو مسارها.")
    print("-" * 30)

# جرب هنا بمسار أي صورة عندك للكتاب (مثلاً سكرين شوت لصفحة من بلاك هورس)
```
